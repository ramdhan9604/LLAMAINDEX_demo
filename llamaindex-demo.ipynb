```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex
from llama_index.core import ServiceContext 
from llama_index.core import StorageContext, load_index_from_storage
```

```python
from llama_index.llms.groq import Groq
```

```python
llm = Groq(model="llama3-70b-8192", api_key=" ")
```

```python
documents = SimpleDirectoryReader("/kaggle/input/yolopdf").load_data()
```

```python
from langchain_community.embeddings.huggingface import HuggingFaceEmbeddings
```

```python
embed_model = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en")
```

```python
from llama_index.core import Settings
```

```python

# Configure settings
Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 800
Settings.chunk_overlap = 20

# Now you can use the Settings object in your code
```

```python
index = VectorStoreIndex.from_documents(documents)
```
